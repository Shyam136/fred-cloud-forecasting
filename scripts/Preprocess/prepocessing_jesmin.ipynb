```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
```

```python
raw_path = "../../data/raw/fred_macro_data_20251029.csv"

df = pd.read_csv(raw_path)

print("✓ Raw data loaded. Shape:", df.shape)
df.head()
```

```python
import numpy as np

df = df.replace([
    ".", "NA", "N/A", "na", "null", "None", "", " "
], np.nan)


if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

print("✓ Cleaned and date parsed")
df.head()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
```
